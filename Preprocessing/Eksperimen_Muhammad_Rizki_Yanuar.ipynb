## **Outliers**

```python
#Check Outliers
plt.figure(figsize=(10, 5))
sns.boxplot(data=stuper, x='Hours_Studied')
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(data=stuper, x='Attendance')
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(data=stuper, x='Sleep_Hours')
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(data=stuper, x='Previous_Scores')
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(data=stuper, x='Tutoring_Sessions')
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(data=stuper, x='Physical_Activity')
plt.show()
```

**ANALISA**

Ditemukan adanya outlier pada 3 atribut numerik:

*  Hours Studied (Mendekati Distribusi Normal)
*  Tutoring Sessions (Right Skewed)

Untuk atribut Tutoring_Sessions, Hours_Studied akan dilakukan teknik capping dimana nilai outlier akan ditarik ke batas tertentu


## **Correlation Matrix**

```python
#Heatmap Correlation
corr = stuper.select_dtypes(include=np.number).corr()
plt.figure(figsize=(10, 5))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

**ANALISA**

Dari hasil korelasi antar variable menunjukkan bahwa, atribut attendance dan Hours studied yang paling berpengaruh dalam hasil ujian akhir (Exam_Score)
